```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
from scipy.optimize import curve_fit

x_data=np.array([-40, 0, 20, 50, 100, 150, 200, 250, 300, 400, 500])
y_data=np.array([1.52, 1.29, 1.20, 1.09, 0.946, 0.835, 0.746, 0.675, 0.616, 0.525, 0.457])

plt.scatter(x_data, y_data)
```

```python
def func(x, a,b,c):
   return a*x**2+b*x+c

popt, pcov=curve_fit(func, x_data, y_data, p0=(0.5, 2, -1))
popt
```

```python
a, b, c = popt

T=np.linspace(-100, 500, 100)
rho=func(T, a, b, c)

plt.scatter(x_data, y_data)
plt.plot(T,rho)
```

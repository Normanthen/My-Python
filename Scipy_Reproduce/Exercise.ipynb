```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
from scipy.optimize import curve_fit

amptd = 1
freq  = 1
x_data=np.linspace(0, 2*np.pi, 50)
y_data=2*amptd*np.sin(freq*x_data)

plt.scatter(x_data, y_data)
```

```python
def func(x,A,w,phi):
    return A*np.sin(w*x+phi)

popt, pcov=curve_fit(func, x_data, y_data, p0=(0.5, 2*np.pi, 0))
popt
```

```python
A, w, phi = popt

t=np.linspace(0, 2*np.pi, 50)
y=func(t, A, w, phi)

plt.scatter(x_data, y_data)
plt.plot(t,y)
```

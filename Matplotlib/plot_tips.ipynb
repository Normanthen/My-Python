Finding max and min value from a plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming x and y are your data points
x = np.linspace(0, 10, 100)  # Replace this with your x data
y = np.sin(x) + 0.1 * np.random.randn(100)  # Replace this with your y data

# Plot the curve
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Curve Plot')
plt.show()

# Find the maximum value and its corresponding x-coordinate
max_y = np.max(y)
max_x = x[np.argmax(y)]

print(f"Maximum value: {max_y} at x = {max_x}")
```

Determining  a value from a specified point in a plot and mark it in the plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Assuming x and y are your data points
x = np.linspace(0, 10, 100)
y = np.sin(x) + 0.1 * np.random.randn(100)

# Plot the curve
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Curve Plot')

# Specify the point where you want to determine the value
desired_x = 5.0

# Find the corresponding y-value
desired_y = np.interp(desired_x, x, y) #interpolate the y-value at the specified x-coordinate

# Mark the point on the graph with an annotation
plt.scatter(desired_x, desired_y, color='red')      #Marking the point
plt.annotate(f'({desired_x:.2f}, {desired_y:.2f})', #add a text annotation at that point
             xy=(desired_x, desired_y),
             xytext=(desired_x, desired_y + 0.5),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             )

plt.show()
```

```python
# Define a coordinate
coordinate = (10, 20)
results = (desired_x,desired_y)

print("The coordinate is:", coordinate)
print('(desired_x,desired_y) = ', results)
```

Creating a Trandline

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Sample data
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 4, 7, 9, 11])

# Define the function for the trendline (e.g., linear)
def linear_function(x, m, b):
    return m * x + b

# Fit the data to the model
params, covariance = curve_fit(linear_function, x_data, y_data)

# Extract the slope and y-intercept from the parameters
slope, intercept = params

# Calculate R-squared value
y_pred = linear_function(x_data, slope, intercept)
ssr = np.sum((y_pred - y_data)**2)
sst = np.sum((y_data - np.mean(y_data))**2)
r_squared = 1 - (ssr / sst)

# Create a line based on the model
trendline = linear_function(x_data, slope, intercept)

# Plot the original data and the trendline
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_data, trendline, label=f'Trendline: y = {slope:.2f}x + {intercept:.2f}\nR-squared: {r_squared:.2f}', color='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

Trendline for 2nd order polynomial

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Sample data
x_data = np.array([1, 2, 3, 4, 5])
y_data = np.array([2, 4, 7, 9, 11])

# Define the function for the 2nd order polynomial trendline
def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

# Fit the data to the model
params, covariance = curve_fit(quadratic_function, x_data, y_data)

# Extract coefficients from the parameters
a, b, c = params

# Calculate R-squared value
y_pred = quadratic_function(x_data, a, b, c)
ssr = np.sum((y_pred - y_data)**2)
sst = np.sum((y_data - np.mean(y_data))**2)
r_squared = 1 - (ssr / sst)

# Create a line based on the model
trendline = quadratic_function(x_data, a, b, c)

# Plot the original data and the trendline
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_data, trendline, label=f'Trendline: y = {a:.2f}x^2 + {b:.2f}x + {c:.2f}\nR-squared: {r_squared:.2f}', color='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

About curve_fit from scipy

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the function to fit
def model_function(x, a, b, c):
    return a * x**2 + b * x + c

# Generate synthetic data with some noise
x_data = np.linspace(0, 10, 100)
y_data = model_function(x_data, 2, -3, 1) + np.random.normal(scale=1, size=len(x_data))

# Use curve_fit to fit the model to the data
fit_params, covariance = curve_fit(model_function, x_data, y_data)

# Extract the fitted parameters
a_fit, b_fit, c_fit = fit_params

# Generate the fitted curve
y_fit = model_function(x_data, a_fit, b_fit, c_fit)

# Plot the original data and the fitted curve
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_data, y_fit, label='Fitted Curve', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Curve Fitting with curve_fit')
plt.show()
```

Cubic spline interpolation

```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 0, 1, 3, 4])

# Create a cubic spline interpolation function
cs = CubicSpline(x, y)

# Generate new x values for interpolation
new_x = np.linspace(1, 5, 100)

# Interpolate y values for the new x values
new_y = cs(new_x)

# Plot the original and interpolated data
plt.scatter(x, y, label='Original Data')
plt.plot(new_x, new_y, label='Cubic Spline Interpolation')
plt.legend()
plt.show()
```

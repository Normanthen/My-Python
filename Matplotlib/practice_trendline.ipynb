```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate linearly spaced data
start_point = 0
end_point = 10
num_points = 100

x_values = np.linspace(start_point, end_point, num_points)

# Define the coefficients of the 2nd order polynomial (ax^2 + bx + c)
a = 2
b = -3
c = 1

# Calculate the corresponding y values for the 2nd order polynomial
y_values = np.polyval([a, b, c], x_values)

# Define the function for the 2nd order polynomial trendline
def quadratic_function(x1, a1, b1, c1):
    return a1 * x1**2 + b1 * x1 + c1

# Fit the data to the model
params, covariance = curve_fit(quadratic_function, x_values, y_values)

# Extract coefficients from the parameters
a1, b1, c1 = params

# Calculate R-squared value
y_pred = quadratic_function(x_values, a1, b1, c1)
ssr = np.sum((y_pred - y_values)**2)
sst = np.sum((y_values - np.mean(y_values))**2)
r_squared = 1 - (ssr / sst)

# Create a line based on the model
trendline = quadratic_function(x_values, a1, b1, c1)

# Plot the trendline
plt.plot(x_values, trendline, label=f'Trendline: y = {a1:.2f}x^2 + {b1:.2f}x + {c1:.2f}\nR-squared: {r_squared:.2f}', color='red')


# Plot the 2nd order polynomial
plt.plot(x_values, y_values, label=f'{a}x^2 + {b}x + {c}')
plt.title('2nd Order Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Sample data
x_data = np.array([20, 30, 40, 50, 60, 75, 100])
y_data = np.array([1.02, 1.53, 2.05, 2.55, 3.07, 3.56, 4.05])

# Define the function for the 2nd order polynomial trendline
def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

# Fit the data to the model
params, covariance = curve_fit(quadratic_function, x_data, y_data)

# Extract coefficients from the parameters
a, b, c = params

# Calculate R-squared value
y_pred = quadratic_function(x_data, a, b, c)
ssr = np.sum((y_pred - y_data)**2)
sst = np.sum((y_data - np.mean(y_data))**2)
r_squared = 1 - (ssr / sst)

# Create a line based on the model
trendline = quadratic_function(x_data, a, b, c)

# Plot the original data and the trendline
plt.scatter(x_data, y_data, label='Original Data')
plt.plot(x_data, trendline, label=f'Trendline: y = {a:.3f}x^2 + {b:.3f}x + {c:.3f}\nR-squared: {r_squared:.2f}', color='red')

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

Calcualting $\mathrm{R^2}$

```python
import numpy as np
from scipy.optimize import curve_fit

# Define the function you want to fit
def func(x, a, b, c):
    return a * x**2 + b * x + c

# Generate some example data
x_data = np.array([1, 2, 3, 4, 5])
y_data = func(x_data, 2, 3, 1) + np.random.normal(scale=0.1, size=len(x_data))  # Add some noise

# Perform the curve fit
params, covariance = curve_fit(func, x_data, y_data)

# Calculate the predicted y values using the fitted parameters
y_fit = func(x_data, *params)

# Calculate the R-squared value
residuals = y_data - y_fit
ss_res = np.sum(residuals**2)
ss_tot = np.sum((y_data - np.mean(y_data))**2)
r_squared = 1 - (ss_res / ss_tot)

print("R-squared:", r_squared)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Generate linearly spaced data
start_point = 0
end_point = 10
num_points = 100

x_values = np.linspace(start_point, end_point, num_points)

# Define the coefficients of the 2nd order polynomial (ax^2 + bx + c)
a = 2
b = -3
c = 1

# Calculate the corresponding y values for the 2nd order polynomial
y_values = np.polyval([a, b, c], x_values)

# Define the function for the 2nd order polynomial trendline
def quadratic_function(x1, a1, b1, c1):
    return a1 * x1**2 + b1 * x1 + c1

# Fit the data to the model
params, covariance = curve_fit(quadratic_function, x_values, y_values)

# Extract coefficients from the parameters
a1, b1, c1 = params

# Calculate R-squared value
y_pred = quadratic_function(x_values, a1, b1, c1)
ssr = np.sum((y_pred - y_values)**2)
sst = np.sum((y_values - np.mean(y_values))**2)
r_squared = 1 - (ssr / sst)

# Create a line based on the model
trendline = quadratic_function(x_values, a1, b1, c1)

# Plot the trendline
plt.plot(x_values, trendline, label=f'Trendline: y = {a1:.2f}x^2 + {b1:.2f}x + {c1:.2f}\nR-squared: {r_squared:.2f}', color='red')


# Plot the 2nd order polynomial
plt.plot(x_values, y_values, label=f'{a}x^2 + {b}x + {c}')
plt.title('2nd Order Polynomial')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
